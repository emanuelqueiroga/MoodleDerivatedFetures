```python
censo['id_moodle']= ''
censo['codigo_persona']= ''
censo['FECHAINGRESOCARR']= ''
censo['fac']= ''
```

```python
for index, row in censo.iterrows():
    numDoc = row['NUMERODOC']
    busca = "NUMERODOC == '" + str(row['NUMERODOC'])+ "'"
    print(busca)
    id_moodle = dadosComplementares.query(busca)[['id']].values
    codigo_persona = dadosComplementares.query(busca)[['codigo_persona']].values
    #DIGITODOC = dadosComplementares.query(busca)[['DIGITODOC']].values
    FECHAINGRESOCARR = dadosComplementares.query(busca)[['FECHAINGRESOCARR']].values
    fac = dadosComplementares.query(busca)[['fac']].values
    censo.loc[index,'id_moodle']= id_moodle
    censo.loc[index,'codigo_persona']= codigo_persona
    censo.loc[index,'FECHAINGRESOCARR']= FECHAINGRESOCARR
    censo.loc[index,'fac']= fac
    print("id")   
    print(id_moodle)
    print(type(id_moodle))
    print("codp")   
    print(codigo_persona)
    print("fecha")   
    print(FECHAINGRESOCARR)
    print("fac")   
    print(fac)
```
